Building the model

```python
from keras import models, layers, regularizers, Input
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# Callbacks
es = EarlyStopping(monitor="val_loss",
                   patience=10,
                   verbose=1,
                   restore_best_weights=True,
                   mode="min")

mc = ModelCheckpoint("best_model.keras",
                     monitor="val_accuracy",
                     verbose=1,
                     save_best_only=True,
                     mode="max")
```

```python
num_of_layers = range(1,4)
num_of_nodes_layer_1 = [32,64,128]
num_of_nodes_layer_2 = [32,64,128]
num_of_nodes_layer_3 = [32,64,128]
activation_functions = ['relu','tanh','sigmoid']
drop_out_rates = [0.3, 0.5]
regularizers_list = ['none','l1','l2']
```

```python
best_val_accuracy = float('-inf')
best_model = None
```

```python
# Helper function to get regularizer
def get_regularizer(reg_type):
    if reg_type == 'none':
        return None
    elif reg_type == 'l1':
        return regularizers.l1(0.1)
    elif reg_type == 'l2':
        return regularizers.l2(0.1)

# Iterate through different configurations
for num_layer in num_of_layers:
    for activation in activation_functions:
        for drop_out in drop_out_rates:
            for regularizer in regularizers_list:
                reg = get_regularizer(regularizer)
                
                if num_layer == 1:
                    for num_node in num_of_nodes_layer_1:
                        input_layer = Input(shape=input_dim)
                        hidden_layer = layers.Dense(num_node, activation=activation, kernel_regularizer=reg)(input_layer)
                        hidden_layer = layers.Dropout(drop_out)(hidden_layer)
                        output_layer = layers.Dense(1, activation='sigmoid')(hidden_layer)
                        model = models.Model(inputs=input_layer, outputs=output_layer)

                elif num_layer == 2:
                    for num_node_1 in num_of_nodes_layer_1:
                        for num_node_2 in num_of_nodes_layer_2:
                            input_layer = Input(shape=input_dim)
                            hidden_layer = layers.Dense(num_node_1, activation=activation, kernel_regularizer=reg)(input_layer)
                            hidden_layer = layers.Dense(num_node_2, activation=activation, kernel_regularizer=reg)(hidden_layer)
                            hidden_layer = layers.Dropout(drop_out)(hidden_layer)
                            output_layer = layers.Dense(1, activation='sigmoid')(hidden_layer)
                            model = models.Model(inputs=input_layer, outputs=output_layer)

                elif num_layer == 3:
                    for num_node_1 in num_of_nodes_layer_1:
                        for num_node_2 in num_of_nodes_layer_2:
                            for num_node_3 in num_of_nodes_layer_3:
                                input_layer = Input(shape=input_dim)
                                hidden_layer = layers.Dense(num_node_1, activation=activation, kernel_regularizer=reg)(input_layer)
                                hidden_layer = layers.Dense(num_node_2, activation=activation, kernel_regularizer=reg)(hidden_layer)
                                hidden_layer = layers.Dense(num_node_3, activation=activation, kernel_regularizer=reg)(hidden_layer)
                                hidden_layer = layers.Dropout(drop_out)(hidden_layer)
                                output_layer = layers.Dense(1, activation='sigmoid')(hidden_layer)
                                model = models.Model(inputs=input_layer, outputs=output_layer)

                # Compile and train the model
                model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
                history = model.fit(
                    X_train, y_train,
                    epochs=20, batch_size=32,
                    validation_data=(X_val, y_val),
                    callbacks=[es, mc], verbose=0
                )
                
                # Evaluate and store the best model
                val_accuracy = max(history.history['val_accuracy'])
                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    best_model = model
```

```python
print("Best Validation Accuracy: ", best_val_accuracy)
```

```python
best_model = keras.models.load_model("best_model.keras")
```
