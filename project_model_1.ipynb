```python
import os
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
```

Creating file path to load the processed data

```python
path = "./data/"
train_file_name = "cleaned_train_data.csv"
test_file_name = "cleaned_test_data.csv"
val_file_name = "cleaned_val_data.csv"

train_fullpath = os.path.join(path,train_file_name)
test_fullpath = os.path.join(path,test_file_name)
val_fullpath = os.path.join(path,val_file_name)
```

Loading the process data


```python
train_data = pd.read_csv(train_fullpath)
train_data
```

```python
test_data = pd.read_csv(train_fullpath)
test_data
```

```python
val_data = pd.read_csv(train_fullpath)
val_data
```

```python
train_data.columns
```

Seperating the X and y

```python
y_train = train_data['FirstYearPersistence']
y_train
```

```python
X_train = train_data.copy()
X_train = X_train.drop('FirstYearPersistence',axis=1)
X_train
```

```python
y_test = test_data['FirstYearPersistence']
y_test
```

```python
X_test = test_data.copy()
X_test = X_test.drop('FirstYearPersistence',axis=1)
X_test
```

```python
y_val = val_data['FirstYearPersistence']
y_val
```

```python
X_val = val_data.copy()
X_val = X_val.drop('FirstYearPersistence',axis=1)
X_val
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# defining the input shape
input_dim = input_dim = (X_train.shape[1],)
input_dim
```

Building the model

```python
from keras import models, layers
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model = models.Sequential()
model.add(layers.Dense(64, activation='relu',input_shape=input_dim))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.3))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.3))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.3))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# Callbacks
es = EarlyStopping(monitor="val_loss",
                   patience=5,
                   verbose=1,
                   mode="min")

mc = ModelCheckpoint("best_model.keras",
                     monitor="val_accuracy",
                     verbose=1,
                     save_best_only=True,
                     mode="max")
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Training
history = model.fit(X_train,
                    y_train,
                    epochs=30,
                    batch_size=32,
                    validation_data=(X_val,y_val),
                    callbacks=[es, mc])
```

Testing the model

```python
test_loss, test_acc = model.evaluate(X_test,y_test, verbose=2)
print(f"Testing accuracy {test_acc}")
```

Quick test to see model preformance

```python
random_index = np.random.randint(0, len(X_test))
sample = X_test.iloc[[random_index]]

prediction = model.predict(sample)

print("Prediction:", prediction[0][0])
print("Actual value:",y_test.iloc[random_index])
```

Saving the model

```python
# saving in 2 formats
model.save("./data/model_1.keras")
model.save("./data/model_1.h5")
```
